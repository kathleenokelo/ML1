#### 2.Develop a strategy to deal with the missing values, i.e deletion, imputation by mean or mode etc, whilst providing rationale for your approach.

```python
# Proportion of missing values
missing_percentage = (chip.isnull().sum() / len(chip)) * 100
print(missing_percentage)
```

For the column 'Release Date' ,impute with mode since it has low percentage of null values

```python
chip['Release Date']=chip['Release Date'].fillna(chip['Release Date'].ffill())
chip['Release Date'].isnull().sum()
```

```python
chip['Release Date'].mode()
```

For the numerical columns 'TDP','Die Size','Transistors','Freq' ,we can use median imputation to maintain distribution

```python
# Convert columns to numeric, setting errors='coerce' to handle non-numeric values
columns_to_convert = ['TDP (W)', 'Die Size (mm^2)', 'Transistors (million)', 'Freq (GHz)']
for col in columns_to_convert:
    chip[col] = pd.to_numeric(chip[col], errors='coerce')
```

```python
chip['TDP (W)'] = chip['TDP (W)'].fillna(chip['TDP (W)'].median())
chip['TDP (W)'].isnull().sum()
```

```python
chip['Die Size (mm^2)'] = chip['Die Size (mm^2)'].fillna(chip['Die Size (mm^2)'].median())
chip['Die Size (mm^2)'].isnull().sum()
```

```python
chip['Transistors (million)'] = chip['Transistors (million)'].fillna(chip['Transistors (million)'].median())
chip['Transistors (million)'].isnull().sum()
```

```python
chip['Freq (GHz)'] = chip['Freq (GHz)'].fillna(chip['Freq (GHz)'].median())
chip['Freq (GHz)'].isnull().sum()
```

For categorical columns such as 'Foundry' we can use mode imputation

```python
chip['Foundry'] = chip['Foundry'].fillna(chip['Foundry'].mode()[0])
chip.isnull().sum()
```

```python
chip['FP32 GFLOPS'].value_counts()
```

Dannard Scaling is still valid in general

Dennard Scaling, formulated by Robert Dennard in 1974, posits that as transistors get smaller, their power density stays constant, meaning power use stays in proportion with area. This scaling implies that voltage and current should scale down with the size of the transistor, which would result in lower power consumption and heat production, allowing for faster clock speeds without increasing power consumption.

```python
# CPUs have higher frequencies, but GPUs are catching up.



# Filter data for CPUs and GPUs
cpu_data = chip[chip['Type'] == 'CPU']
gpu_data = chip[chip['Type'] == 'GPU']

# Aggregate by year and compute mean frequency
cpu_yearly = cpu_data.groupby('Year')['Freq (GHz)'].mean().reset_index()
gpu_yearly = gpu_data.groupby('Year')['Freq (GHz)'].mean().reset_index()

# Plotting
fig,ax = plt.subplots(figsize=(14, 8))
ax.plot(cpu_yearly['Year'], cpu_yearly['Freq (GHz)'], label='CPU', marker='o')
ax.plot(gpu_yearly['Year'], gpu_yearly['Freq (GHz)'], label='GPU', marker='o')
ax.set_title('Mean Frequency of CPUs and GPUs Over Time')
ax.set_xlabel('Year')
ax.set_ylabel('Frequency (GHz)')
ax.set_xlim(1998,2025)
ax.locator_params(axis='x',nbins=27)
plt.legend()
plt.grid(True)
plt.show()
```

One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

def column_names(feature,category):
    return str(feature)+'-'+str(category)

ohe=OneHotEncoder(feature_name_combiner=column_names)

ohe_values = ohe.fit_transform(chip[['Type']]).toarray()
columns = ohe.get_feature_names_out()

ohe_enc = pd.DataFrame(ohe_values,columns=columns)
chip=pd.concat([ohe_enc,chip],axis=1).drop(columns=['Type-CPU'])

```
